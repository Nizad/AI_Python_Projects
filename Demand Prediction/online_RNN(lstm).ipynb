```
class Online_RNN():
  def __init__(self):
    print('this is An online algorithem')
  def start(self):
    path = input('please write your data path: ')
    cf= pd.read_excel(f'{path}')
    i = 6
    j = 0
    boundary = 0.4
    all_scores = []
    all_mse=[]
    while i <100:
      train = cf.iloc[:,i-5:i]
      target = cf.iloc[:,[i+1,i+2]]
      train = np.array(train)
      target = np.array(target)
      X =train
      n_features = 1
      X = X.reshape((X.shape[0], X.shape[1], n_features))
      model = Sequential()
      model.add(Bidirectional(LSTM(100, activation='relu'), input_shape=(5, n_features)))
      model.add(Dense(2))
      model.compile(optimizer='adam', loss='mse',metrics=['accuracy'])
      history1 =model.fit(X, target, epochs=10, validation_split=0.3,verbose=2)
      val_mse, val_mae = model.evaluate(X, target, verbose=0)
  
      all_scores.append(val_mae)
      all_mse.append(val_mse)
      print(f'the final accuracy is : {all_scores[j]}')
      print(f'the final loss is : {all_mse[j]}')
      print(f"this is the{i-5} to {i} columns")
      print(f"this is the {i+1} and {i+2} columns for target")
      pd.DataFrame(history1.history).plot(figsize=(5, 5))
      plt.grid(True)
      plt.gca().set_xlim(0, 10)
      plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
      plt.xlabel('number of epochs', fontsize=14)
      plt.ylabel('accuracy and loss', fontsize=14)
      plt.title(f'online model for prediction the{i-5} to {i} columns as input  ',fontsize=16) 
      plt.show()
      if all_mse[j]<boundary:
        i+=1
      else:
        print('the error band has been reached')
        print(f'the erroe is {all_mse[-1]}which is more than our {boundary} boundary')
        break
      j+=1  
  
```

```
onlin = Online_RNN()
onlin.start()
```
